```python
# Both Numeric and Non numeric - groupby() function is used
# Leftside of groupby must be numerical
# Rightside of groupby within brackets Non numeric
# Statistical Function like,sum,mean,median,standard deviation,etc must be specified
```

```python
# Total number_of_reviews for top 10 neighbourhoods
Chicago.number_of_reviews.groupby(
    Chicago.neighbourhood).sum().sort_values(
    ascending=False).nlargest(10)
```

```python
ax=Chicago.number_of_reviews.groupby(
    Chicago.neighbourhood).sum().sort_values(
    ascending=False).nlargest(10).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Average price of Top 20 Neighbourhood
Chicago.price.groupby(
Chicago.neighbourhood).mean().sort_values(
ascending=False).nlargest(20)
```

```python
plt.figure(figsize=(15,7))
ax=Chicago.price.groupby(
    Chicago.neighbourhood).mean().sort_values(
    ascending=False).nlargest(20).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize=(15,7))
ax=Chicago.price.groupby(
    Chicago.neighbourhood).mean().sort_values(
    ascending=False).nsmallest(20).plot(kind='bar',color='green')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Mean -(N)/N
# Median - Sort data in ascending order,even (m1+m2)/2,odd-exact Midpoint
# Mode - Most Frequent or repetitive

# Variance - (x-mean)^2/N-1  N - Population & N-1 - Sample
# Standard Deviation - sqrt(variance)

# Skewness - (x-mean)^3/stdev^3
# Kurtosis - (x-mean)^4/stdev^4

# Quartiles - Sort Data in ascending Order, Q1 - N/4, Q2-2N/4, Q3-3N/4
```
